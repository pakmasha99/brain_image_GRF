# 정답을 찾음:
tldr : `adjust_learning_rate`함수에서 볼 수 있듯이, learing rate를 지정해줄때 max_steps를 기준으로 해서 거기를 최대일떄로 해서서 거기서부부l learning rate를 지정해주는데 (밑의 그림 보기), 그런데 내가 돌리는 중간에 epochs를 바꾸면 다르게 준다! (즉, 밑의 그림에서 lr이 0에 가까워지는 시기가 main넣어줄때 epoch값을 다르게 하면 다르게 되기때문에 값이 이상해 지는 것 )


**밑 : 그것을 설명하는 것 **


```python
def adjust_learning_rate(args, optimizer, loader, step):
    max_steps = args.epochs * len(loader)
    warmup_steps = 10 * len(loader)
    base_lr = args.batch_size / 256
    if step < warmup_steps:
        lr = base_lr * step / warmup_steps
    else:
        step -= warmup_steps
        max_steps -= warmup_steps
        q = 0.5 * (1 + math.cos(math.pi * step / max_steps))
        end_lr = base_lr * 0.001
        lr = base_lr * q + end_lr * (1 - q)
    optimizer.param_groups[0]['lr'] = lr * args.learning_rate_weights
    optimizer.param_groups[1]['lr'] = lr * args.learning_rate_biases

```

이렇게 해서 optimizer의 lr_weight, lr_bias를 지정해준다 (using param_groups to update the bias and weights with seperate parametger configs)


그런데, 
`optimizer.param_groups[0]['lr'] = lr * args.learning_rate_weights`에서, `args.learning_rate_weights`는 이미 지정된 값이다 (same regardless of resuming the thing or not)]

> **따라서, `lr` 이 이상하게 되어서 resume할때 값이 다른 듯하다!**

lr은 다양 한 것들에 의해서 정의가 된다
* `step` : 이거는 안전 (`for step, ((y1, y2), _) in enumerate(loader, start=epoch * len(loader)):` 에서 start 를 지정해주니)
* `max_steps`: 이것도 안전 (because `args.epochs`, `len(loader)`은 이미 정해진 값들 )

문제해결하려고, 똑같은 것을 더 작은 dataset가져와서 돌렸다가 끊어보고 해서 lr의 변화를 봄 (`adjust learning rate`함수의 lr값등을 print하도록 함.. 변화를 보려고)


만약 저 상태에서 epoch을 안바꾸고 다시 돌리면, (i.e. restarts from epoch 1)

위에서와 같이, lr, lr_biases등을 잘 이어받았다는 것을 알 수 있따 

단, python script의 `--epochs` 값을 바꾸면 달라진다! (밑에 보기)(4에서 8로 바꿈) (여전히 epoch = 1에서 시작)(not in the middel of epoch 1.5 or sth)




아 warmup때는 괜찮겠다 (dataloader 의 length (i.e. number of batches)가 변하지 않는 이상)(due to the definition of "adjust_learning_rate" 함수)

````python
from torch import nn, optim
import torch
import torchvision
import torchvision.transforms as transforms
````

````python
class Args():
    def __init__(self):
        self.epochs = 30
        self.batch_size = 40
        self.projector = '8192-8192-8192'
        self.weight_decay = 1e-1
        self.lambd = 0.0051
        self.learning_rate_weights = 0.0048
        self.learning_rate_biases = 0.2
args = Args()


gpu = 'cuda:0'
````

````python
class LARS(optim.Optimizer):
    def __init__(self, params, lr, weight_decay=0, momentum=0.9, eta=0.001,
                 weight_decay_filter=False, lars_adaptation_filter=False):
        defaults = dict(lr=lr, weight_decay=weight_decay, momentum=momentum,
                        eta=eta, weight_decay_filter=weight_decay_filter,
                        lars_adaptation_filter=lars_adaptation_filter)
        super().__init__(params, defaults)


    def exclude_bias_and_norm(self, p):
        return p.ndim == 1

    @torch.no_grad()
    def step(self):
        for g in self.param_groups:
            for p in g['params']:
                dp = p.grad

                if dp is None:
                    continue

                if not g['weight_decay_filter'] or not self.exclude_bias_and_norm(p):
                    dp = dp.add(p, alpha=g['weight_decay'])

                if not g['lars_adaptation_filter'] or not self.exclude_bias_and_norm(p):
                    param_norm = torch.norm(p)
                    update_norm = torch.norm(dp)
                    one = torch.ones_like(param_norm)
                    q = torch.where(param_norm > 0.,
                                    torch.where(update_norm > 0,
                                                (g['eta'] * param_norm / update_norm), one), one)
                    dp = dp.mul(q)

                param_state = self.state[p]
                if 'mu' not in param_state:
                    param_state['mu'] = torch.zeros_like(p)
                mu = param_state['mu']
                mu.mul_(g['momentum']).add_(dp)

                p.add_(mu, alpha=-g['lr'])
````

````python
class BarlowTwins(nn.Module):
    def __init__(self, args):
        super().__init__()
        self.args = args
        self.backbone = torchvision.models.resnet50(zero_init_residual=True)
        self.backbone.fc = nn.Identity()

        # projector
        sizes = [2048] + list(map(int, args.projector.split('-')))
        layers = []
        for i in range(len(sizes) - 2):
            layers.append(nn.Linear(sizes[i], sizes[i + 1], bias=False))
            layers.append(nn.BatchNorm1d(sizes[i + 1]))
            layers.append(nn.ReLU(inplace=True))
        layers.append(nn.Linear(sizes[-2], sizes[-1], bias=False))
        self.projector = nn.Sequential(*layers)

        # normalization layer for the representations z1 and z2
        self.bn = nn.BatchNorm1d(sizes[-1], affine=False)

    def forward(self, y1, y2):
        z1 = self.projector(self.backbone(y1))
        z2 = self.projector(self.backbone(y2))

        # empirical cross-correlation matrix
        c = self.bn(z1).T @ self.bn(z2)

        # sum the cross-correlation matrix between all gpus
        c.div_(self.args.batch_size)
        torch.distributed.all_reduce(c)

        on_diag = torch.diagonal(c).add_(-1).pow_(2).sum()
        off_diag = off_diagonal(c).pow_(2).sum()
        loss = on_diag + self.args.lambd * off_diag
        return loss
````

````python

model = BarlowTwins(args).cuda(gpu)
#model = nn.SyncBatchNorm.convert_sync_batchnorm(model)
param_weights = []
param_biases = []
for param in model.parameters():
    if param.ndim == 1:
        param_biases.append(param)
    else:
        param_weights.append(param)
parameters = [{'params': param_weights}, {'params': param_biases}]
#model = torch.nn.parallel.DistributedDataParallel(model, device_ids=[gpu])
optimizer = LARS(parameters, lr=0, weight_decay=args.weight_decay,
                 weight_decay_filter=True,
                 lars_adaptation_filter=True)
````

````python
def adjust_learning_rate(args, optimizer, loader, step):
    loader = [2,3,4]
    max_steps = args.epochs * len(loader)
    warmup_steps = 10 * len(loader)
    base_lr = args.batch_size / 256
    if step < warmup_steps:
        lr = base_lr * step / warmup_steps
    else:
        step -= warmup_steps
        max_steps -= warmup_steps
        q = 0.5 * (1 + math.cos(math.pi * step / max_steps))
        end_lr = base_lr * 0.001
        lr = base_lr * q + end_lr * (1 - q)
    optimizer.param_groups[0]['lr'] = lr * args.learning_rate_weights
    optimizer.param_groups[1]['lr'] = lr * args.learning_rate_biases

    ##added for testing
    return lr
````
