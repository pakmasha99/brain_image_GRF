```python
import pandas as pd
import numpy as np
import re

####################### 필요한 함수 준비 #########################
##내가 만든 내가 자주 사용하느 함수
#1) 자료 이름 바꾸기_ subject name 에 언더바나 이런게 있는 경우에 인식이 안되니까
def change_name (df):
  old = []
  new = []
  for i in df["subjectkey"]:
    old.append(i)
  
  for i in old:
    i = re.sub(r'_', '', i)
    new.append(i)

  df1 = df
  df1["subjectkey"] = new
  return df1

#2) scale 함수 
#scaling 하기
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

def scale(df):
  train = df

  train_except = train.drop(["subjectkey"],axis= 1)
  name = train["subjectkey"].values
  fitted = scaler.fit(train_except)
  train_scaled = scaler.transform(train_except)
  train_output = pd.DataFrame(train_scaled, columns = train_except.columns, index=list(train_except.index.values))
  train_output.insert(0,"subjectkey",name)
  df2 = train_output

  print('scale 전 feature들의 분산 값')
  print(df.var())
  print('scale 후 feature들의 분산 값')
  print(df2.var())

  return df2

#3특정 열에서 oberservation이 1000개 미만이면 버리기
#strucure & fa에서 왕창 다 0인 것들 모아서 살려내기
#1) struc에서 0이 7000개 이상인 것들은 lesion관련한 건데 (당연함.애들임)
def dropcol(df):
  listy = list(df.columns)
  for i in listy:
    if df[i].count() < 1000:
      df.drop(i,axis = 1,inplace = True)
  return df
```

```python
def count_1(df):
  s = 0 
  for i in df.index:
    if df[i] == 1:
      s+=1
  return s

def count_0(df):
  s = 0
  for i in df.index:
    if df[i] == 0:
      s+=1
  return s


def count_2(df):
  s = 0
  for i in df.index:
    if df[i] == 2:
      s+=1
  return s
```

```python
#dataload
exclu = pd.read_csv("/content/drive/MyDrive/ABCD_phenotype_total.csv")
ksads_C = pd.read_csv("/content/drive/MyDrive/new_ksads_v2_child.csv")
ksads_P = pd.read_csv("/content/drive/MyDrive/new_ksads_v2_parent.csv")
```

```python
exclu = change_name(exclu)
ksads_C = change_name(ksads_C)
ksads_P = change_name(ksads_P)
```

```python
ksads_C.drop(["eventname"],axis =1, inplace = True)
ksads_merged = pd.merge(ksads_C , ksads_P, how = "inner", on = "subjectkey")
ksads_merged 
```

```python
# Ksads_ ADHD 열 만들기 
#기준 1 = ksads 부모랑 애들 거기서 다 0이어야 함 (no suic 제외 )
#HC_noksads를 만들기 위해서 먼저 merge 
df_merged = ksads_merged 
name = df_merged["subjectkey"]
df_merged2 = df_merged.loc[:,df_merged.columns!='subjectkey']
listy = []
for i in df_merged2.sum(axis =1):
  listy.append(i)

df_merged2.insert(0,"subjectkey",name)
df_merged2.insert(1,"KSADS_sum",listy)

noksads = []
for i in df_merged2.index:
  if df_merged2["KSADS_sum"][i] == 0:
    noksads.append(0)
  else:
    noksads.append(1)

df_merged2.insert(1,"HC_noKSADS",noksads)
df_merged2= df_merged2.drop("KSADS_sum",axis= 1)
df_merged2
```

```python
cbcl = pd.read_csv("/content/drive/MyDrive/cbcl_v2.csv")
cbcl_real = change_name(cbcl)
cbcl_real.drop("eventname",axis =1 , inplace = True)

#covariates
cov_real = exclu.loc[:,["subjectkey","sex"]] #"eventname"
```

```python
cbcl_real.columns
```

```python
#cbcl_scr_dsm5_adhd_t
#CBCL_ADHD열 만들기
# 기준 2 - CBCL 60보다 적어야 함
#HC_noCBCL을 만들기 위해서 먼저 merge 
df_merged_C = pd.merge(df_merged2,cbcl_real, how = "inner", on = "subjectkey") 
name = df_merged_C["subjectkey"]
df_merged3 = df_merged_C.loc[:,df_merged_C.columns!='subjectkey']
merged_cbcl = df_merged3.loc[:,"cbcl_scr_syn_anxdep_t":"cbcl_scr_07_stress_t"]# [:,"cbcl_scr_syn_anxdep_t":"cbcl_scr_07_stress_t"][:,"AnxDep.CBCL":"Stress.CBCL"]

listy2 = []#여기가 원래 set()이었는데 왜 list이면 안되는거지? 아래로 가면 나오는데 말이지? 

for i in merged_cbcl.index:
  for j in merged_cbcl.columns:
    if merged_cbcl[j][i]>= 60:
      listy2.append(i)

#listy2 = list(listy2)

nocbcl = []

for i in df_merged3.index:
  if i in listy2:
    nocbcl.append(1)
  else: 
    nocbcl.append(0)

df_merged3.insert(0,"subjectkey",name)
df_merged3.insert(1,"HC_noCBCL",nocbcl)
df_merged3    
```

```python
# 마지막으로 cov_real 합쳐줘야지 
#merge하기 #
df_merged = pd.merge( df_merged3  ,cov_real, how = "inner", on = "subjectkey")
#df_merged = pd.merge( df_merged ,cbcl_real, how = "inner", on = "subjectkey")
df_merged.dropna(inplace = True)
df_merged.isnull().sum().sum()
```
