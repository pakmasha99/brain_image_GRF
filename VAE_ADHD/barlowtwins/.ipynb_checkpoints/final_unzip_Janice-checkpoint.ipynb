# 분리하기 

```python
data_path = '/scratch/connectome/mieuxmin/UKB_t1_MNI/zip_original/'
```

```python
file_list = os.listdir(data_path) 
```

```python
#file_list
```

```python
#from zipfile import ZipFile
#import shutil

for i in file_list:
    number = i[:7]
    i = data_path + i
    path_to_zip_file = i
    directory_to_extract_to = "/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_1000"
    with ZipFile(path_to_zip_file, 'r') as zipObj:
        listOfFileNames = zipObj.namelist()
        for fileName in listOfFileNames:
            if fileName.endswith("T1_brain_to_MNI.nii.gz"):
                zipObj.extract(fileName, "/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_mni_1000/"+number+"_T1_brain_to_MNI")
                os.chdir("/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_mni_1000/"+number+"_T1_brain_to_MNI/T1")
                os.getcwd()
                os.rename("T1_brain_to_MNI.nii.gz", number+"_T1_brain_to_MNI.nii.gz")
                file = number+"_T1_brain_to_MNI.nii.gz"
                src = "/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_mni_1000/"+number+"_T1_brain_to_MNI/T1/"
                dir = "/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_mni_1000/"
                shutil.move(src+file , dir+file)
                shutil.rmtree("/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_mni_1000/"+number+"_T1_brain_to_MNI/")
                
            elif fileName.endswith("T1_brain.nii.gz"):
                zipObj.extract(fileName, "/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_raw_1000/"+number+"brain")
                os.chdir("/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_raw_1000/"+number+"brain/T1")
                os.getcwd()
                os.rename("T1_brain.nii.gz", number+"_T1_brain.nii.gz")
                file = number+"_T1_brain.nii.gz"
                src = "/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_raw_1000/"+number+"brain/T1/"
                dir = "/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_raw_1000/"
                shutil.move(src+file , dir+file)
                shutil.rmtree("/scratch/connectome/mieuxmin/UKB_t1_MNI/unzip_raw_1000/"+number+"brain/")
                
            else:
                pass
print("완료")
```
