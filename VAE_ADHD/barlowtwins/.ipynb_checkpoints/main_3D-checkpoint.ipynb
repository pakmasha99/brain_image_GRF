### main_worker 정의
`torch.multiprocessing.spawn(main_worker, ...)`이라고 불러오는 것으로 봐서, DDP의 어떤 command 인듯?

````python
#example : `cuda:0` ?? 
#
def main_worker(gpu, args):
    args.rank += gpu   #i.e. rank = rank + gpu(왜 이렇게 하는지는 모르겠다....)
    torch.distributed.init_process_group(
        backend='nccl', init_method=args.dist_url,
        world_size=args.world_size, rank=args.rank)

    if args.rank == 0: #i.e. if gpu =0 and (원래) args.rank = 0 (?)
        args.checkpoint_dir.mkdir(parents=True, exist_ok=True)
        stats_file = open(args.checkpoint_dir / 'stats.txt', 'a', buffering=1)
        print(' '.join(sys.argv))
        print(' '.join(sys.argv), file=stats_file)

    torch.cuda.set_device(gpu)
    torch.backends.cudnn.benchmark = True

    model = BarlowTwins(args).cuda(gpu)
    model = nn.SyncBatchNorm.convert_sync_batchnorm(model)
    param_weights = []
    param_biases = []
    for param in model.parameters():
        if param.ndim == 1:
            param_biases.append(param)
        else:
            param_weights.append(param)
    parameters = [{'params': param_weights}, {'params': param_biases}]
    model = torch.nn.parallel.DistributedDataParallel(model, device_ids=[gpu])
    optimizer = LARS(parameters, lr=0, weight_decay=args.weight_decay,
                     weight_decay_filter=True,
                     lars_adaptation_filter=True)

    # automatically resume from checkpoint if it exists
    if (args.checkpoint_dir / 'checkpoint.pth').is_file():
        ckpt = torch.load(args.checkpoint_dir / 'checkpoint.pth',
                          map_location='cpu')
        start_epoch = ckpt['epoch']
        model.load_state_dict(ckpt['model'])
        optimizer.load_state_dict(ckpt['optimizer'])
    else:
        start_epoch = 0

    dataset = torchvision.datasets.ImageFolder(args.data / 'train', Transform())
    sampler = torch.utils.data.distributed.DistributedSampler(dataset)
    assert args.batch_size % args.world_size == 0
    per_device_batch_size = args.batch_size // args.world_size
    loader = torch.utils.data.DataLoader(
        dataset, batch_size=per_device_batch_size, num_workers=args.workers,
        pin_memory=True, sampler=sampler)

    start_time = time.time()
    scaler = torch.cuda.amp.GradScaler()
    for epoch in range(start_epoch, args.epochs):
        sampler.set_epoch(epoch)
        for step, ((y1, y2), _) in enumerate(loader, start=epoch * len(loader)):
            y1 = y1.cuda(gpu, non_blocking=True)
            y2 = y2.cuda(gpu, non_blocking=True)
            adjust_learning_rate(args, optimizer, loader, step)
            optimizer.zero_grad()
            with torch.cuda.amp.autocast():
                loss = model.forward(y1, y2)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()
            if step % args.print_freq == 0:
                if args.rank == 0:
                    stats = dict(epoch=epoch, step=step,
                                 lr_weights=optimizer.param_groups[0]['lr'],
                                 lr_biases=optimizer.param_groups[1]['lr'],
                                 loss=loss.item(),
                                 time=int(time.time() - start_time))
                    print(json.dumps(stats))
                    print(json.dumps(stats), file=stats_file)
        if args.rank == 0:
            # save checkpoint
            state = dict(epoch=epoch + 1, model=model.state_dict(),
                         optimizer=optimizer.state_dict())
            torch.save(state, args.checkpoint_dir / 'checkpoint.pth')
    if args.rank == 0:
        # save final model
        torch.save(model.module.backbone.state_dict(),
                   args.checkpoint_dir / 'resnet50.pth')
````

````python
class GaussianBlur(object):
    def __init__(self, p):
        self.p = p

    def __call__(self, img):
        if random.random() < self.p:
            sigma = random.random() * 1.9 + 0.1
            return img.filter(ImageFilter.GaussianBlur(sigma))
        else:
            return img


class Solarization(object):
    def __init__(self, p):
        self.p = p

    def __call__(self, img):
        if random.random() < self.p:
            return ImageOps.solarize(img)
        else:
            return img
````

````python
class Transform:
    def __init__(self):
        self.transform = transforms.Compose([
            transforms.RandomResizedCrop(224, interpolation=Image.BICUBIC),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomApply(
                [transforms.ColorJitter(brightness=0.4, contrast=0.4,
                                        saturation=0.2, hue=0.1)],
                p=0.8
            ),
            transforms.RandomGrayscale(p=0.2),
            GaussianBlur(p=1.0),
            Solarization(p=0.0),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
        ])
        self.transform_prime = transforms.Compose([
            transforms.RandomResizedCrop(224, interpolation=Image.BICUBIC),
            transforms.RandomHorizontalFlip(p=0.5),
            transforms.RandomApply(
                [transforms.ColorJitter(brightness=0.4, contrast=0.4,
                                        saturation=0.2, hue=0.1)],
                p=0.8
            ),
            transforms.RandomGrayscale(p=0.2),
            GaussianBlur(p=0.1),
            Solarization(p=0.2),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
        ])

    def __call__(self, x):
        y1 = self.transform(x)
        y2 = self.transform_prime(x)
        return y1, y2



````

## X . model 정의

> 전체적인 model structure
1. Resnet 50 (참조 : (https://arxiv.org/pdf/1512.03385.pdf), (https://jisuhan.tistory.com/71))
    * 구조 : INPUT => CNN (resnet) =(flatten)=> 2048 =(FC)=> 1000 (labels)
        * 50 layer resnet
        * input : (3, H, W), with pixel within [0,1]
        * output : 1000 output vector (labels)
    * 여기서는, 2048 => 1000을 안하고, 바로 2048 => 2048로 되도록 backbone.fc 를 nn.Identity()로 바꿈!
2. 
ㅇ
ㅇ
ㅇ


````python
class BarlowTwins(nn.Module):
    def __init__(self, args):
        super().__init__()
        self.args = args
        self.backbone = torchvision.models.resnet50(zero_init_residual=True)
        self.backbone.fc = nn.Identity() #1000이아닌 2048 vector output!

        # projector             (결과물 : (barlow_model.projector 로 보기 가능))
        sizes = [2048] + list(map(int, args.projector.split('-'))) # list : [2048, 8192, 8192, 8192]
        layers = []
        for i in range(len(sizes) - 2):                             #i.e. 2048 -> 8192 -> 8192 -> 8192 사이에 nn.Linear, Batchnorm, ReLU 하기!
            layers.append(nn.Linear(sizes[i], sizes[i + 1], bias=False))
            layers.append(nn.BatchNorm1d(sizes[i + 1]))
            layers.append(nn.ReLU(inplace=True))
        layers.append(nn.Linear(sizes[-2], sizes[-1], bias=False))
        self.projector = nn.Sequential(*layers)

        # normalization layer for the representations z1 and z2
        self.bn = nn.BatchNorm1d(sizes[-1], affine=False) #affine : false (because 실제 'normalization'만 해야해서)

    def forward(self, y1, y2):
        z1 = self.projector(self.backbone(y1))           #i.e. z1 : representation of y1 (before normalization)
        z2 = self.projector(self.backbone(y2))           #i.e. z2 : representation of y2 (before normalization)

        # empirical cross-correlation matrix
        c = self.bn(z1).T @ self.bn(z2)                  #즉, normalized된 z1, z2를 outer product 한다 (cross correlation 구하기)
                    #(dimensino in the middle is the number of batches, wihch is the number of samples in this case) 

        # sum the cross-correlation matrix between all gpus    #DDP 공부하고 하기!
        c.div_(self.args.batch_size)
        torch.distributed.all_reduce(c)
        
        #calculate loss(?) (이 파트 밑에 supplementary로 적는 것 있으니 하기!)
        on_diag = torch.diagonal(c).add_(-1).pow_(2).sum()
        off_diag = off_diagonal(c).pow_(2).sum()
        loss = on_diag + self.args.lambd * off_diag 
        return loss
````

#### X-1-3. Batchnorm 파트 (for z1, z2 normalization) 파트 : 
Batchnorm 이 뭔지 : https://wegonnamakeit.tistory.com/47





**X-1-3-1.** 밑에서 : batchnorm이 뭔지 밑에 예시로 보여줌, 그리고 batchnorm 직접 계산한거랑 nn.Batchnorm1d 써서 계산한거랑 같다는 것을 보임!

* 여기서 주의할 점 : `unbiased = False`로 했다! 이 뜻은 무엇이냐면, population average 와 sample avearge가 비슷하도록 Bessel Correction을 $\sqrt{\frac{n}{n-1}}$로 해줘야하는데, 그것을 안햇다는 것! 
    * 따라서, `a.std`구할때 `unbiased = False`로 한것! (i.e. no Bessel Correction)
    * 안한 이유 : batchnorm은 "실제" layer-wise std 가 1로 standardize 되어야해서! 
    * 관련 문헌 : 
        * https://pytorch.org/docs/stable/generated/torch.std.html
        * https://en.wikipedia.org/wiki/Bessel%27s_correction




````python
a = torch.randn((3,10))

#true batchnorm result
true_result = nn.BatchNorm1d(10)(a)

#expected batchnorm result
a_mean = a.mean(axis = 0)
a_std = a.std(axis = 0, unbiased = False)
expected_result = (a-a_mean)/a_std
print("a :\n",a)
print("\ntrue batchnorm result of a :\n", true_result)
print("\nexpected batchnorm result of a :\n",expected_result)


print("\nthe ratio between expected_result and true_result :\n", expected_result/true_result)
#1, as expected


#실제로 layer-wise로 평균이 0, std가 1인지 보자
true_result.mean(axis=0), true_result.std(axis=0, unbiased = False)
````

**X-1-3-2.** `affine = False`이다!
위에서 barlow twins정의할때, 거의 모든 batchnorm 은 affine언급이 없는데 (i.e. default : True), z1, z2를 담당하는 `self.bn`에서만, `affine = False`를 한다!

그 이유 : 밑의 batchnorm documentation을 보면 알 수 있듯이, 원래는 batchnorm시 learnable parameter로 $\gamma, \beta$를 가지게 되는데, 우리는 z1, z2를 normalize할때, "진짜로" noramlize**만**해야한다 (즉, $\gamma, \beta$ 값들이 1,0이 아니게 된다면, 실제 average와 std가 0,1이 아니가 되어, normalizatoin "만" 하는게 아니게됨!!!) 따라서, normalization만 해야해서 `affine = False`로 하는 것!


batchnorm documentation : 

https://pytorch.org/docs/stable/generated/torch.nn.BatchNorm1d.html


````python
#affine = True (default)
print(nn.BatchNorm1d(num_features=10, affine = True)(a))

#affine = False
print(nn.BatchNorm1d(num_features=10, affine = False)(a))
````

### 2.1. 고친 densnet으로 `BarlowTwins` 모델 만들기!


````python
class BarlowTwins(nn.Module):
    def __init__(self, args):
        super().__init__()
        self.args = args
        
        #======바뀐부분 : densnet을 가져옴 (instead of the resnet50====#
        self.backbone = densenet121(mode = "CNN_flatten", drop_rate = 0.0)
        ##
        
        #self.backbone.fc = nn.Identity() #1000이아닌 2048 vector output!

        # projector             (결과물 : (barlow_model.projector 로 보기 가능))
        #======바뀐부분 : 밑에 2048에서 1024로 바뀜 (densenet shape맞추기 위해서)====#
        sizes = [1024] + list(map(int, args.projector.split('-'))) # list : [2048, 8192, 8192, 8192]
        layers = []
        for i in range(len(sizes) - 2):                             #i.e. 2048 -> 8192 -> 8192 -> 8192 사이에 nn.Linear, Batchnorm, ReLU 하기!
            layers.append(nn.Linear(sizes[i], sizes[i + 1], bias=False))
            layers.append(nn.BatchNorm1d(sizes[i + 1]))
            layers.append(nn.ReLU(inplace=True))
        layers.append(nn.Linear(sizes[-2], sizes[-1], bias=False))
        self.projector = nn.Sequential(*layers)

        # normalization layer for the representations z1 and z2
        self.bn = nn.BatchNorm1d(sizes[-1], affine=False) #affine : false (because 실제 'normalization'만 해야해서)

    def forward(self, y1, y2):
        z1 = self.projector(self.backbone(y1))           #i.e. z1 : representation of y1 (before normalization)
        z2 = self.projector(self.backbone(y2))           #i.e. z2 : representation of y2 (before normalization)

        # empirical cross-correlation matrix
        c = self.bn(z1).T @ self.bn(z2)                  #즉, normalized된 z1, z2를 outer product 한다 (cross correlation 구하기)
                    #(dimensino in the middle is the number of batches, wihch is the number of samples in this case) 

        # sum the cross-correlation matrix between all gpus    #DDP 공부하고 하기!
        c.div_(self.args.batch_size)
        torch.distributed.all_reduce(c)
        
        #calculate loss(?) (이 파트 밑에 supplementary로 적는 것 있으니 하기!)
        on_diag = torch.diagonal(c).add_(-1).pow_(2).sum()
        off_diag = off_diagonal(c).pow_(2).sum()
        loss = on_diag + self.args.lambd * off_diag 
        return loss

````

````python
#barlow model structure어떻게 되는지 보기
barlow_model = BarlowTwins(args)
#print(barlow_model) #sum of .backbone + .projector + XXX
#print(barlow_model.backbone) #densenet part 
print(barlow_model.projector) #projector part
print(barlow_model.bn)        #z1, z2 (representation)의 batchnormalization
````
