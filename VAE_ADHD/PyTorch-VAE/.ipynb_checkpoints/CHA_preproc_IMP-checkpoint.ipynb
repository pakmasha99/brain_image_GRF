## Step1 : age 를 계산하는 (and removing NaN) 것을 만들기 
(밑에 : 예시로 2021만 함.. 나머지는 .py해서 돌리던지 하기)

```python
file = pd.read_excel(tabular_path+'2021년 목록(최종) - 총 72개, 4개 missing.xlsx')
print(file.shape) #file name은 위의 os.listdir결과들 쓰기
file.head()
```

```python
##remove 파일 NaN
file = file.dropna(subset=['파일 유무'])  #excel에서 원래 hiding함
            #==O를 쓰지 않은 이유 : o을 쓴 경우도 있어서 ㅋㅋㅋ

##remove 파일유무 = X 
#일단 이거는 생략 (MRI 둘다 없는 아이들 따로 하니) => DTI없는 아이들한테 freesurfer만 해야할 수도 있으니 살려둠

##get age
mri_date = file['MRI(DTI) 시행 날짜']
birth_date = file['Date of birth']
#change age to pandas dtype : datetime64 (to get change in dates to get age)
mri_date = pd.to_datetime(mri_date)
birth_date = pd.to_datetime(birth_date)

age_days = mri_date - birth_date #age in days
age_years = age_days/np.timedelta64(1, 'Y')


##append age_years to the dataset
file['age'] = age_years

file.head() #for checxking
```

## Step2 : age 4.5기준으로 나누기

```python
#plt.hist(file['age(y)'], bins=10) #age 분포 보기
infant_mask = file['age'] < 4.5
adult_mask = file['age'] >=4.5
##pediatric 써야하면 조금 수정을 가해야할수도
##일단은 이렇게 까지 해놨으니 될듯
```
