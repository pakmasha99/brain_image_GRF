```python
phenotype = pd.read_csv("/scratch/connectome/mieuxmin/UKB_t1_MNI/UKB_phenotype.csv") 
phenotype_real = phenotype[["eid","sex"]]
phenotype_real = phenotype_real.astype({"eid":'str'})
phenotype_real
```

```python
# val group과 train group을 위한 phenotype csv가 필요함
mode = "val"
print(mode)
data_path = "/scratch/connectome/mieuxmin/UKB_t1_MNI/npy_val_unzip_mni_100" #mode에따라서 바꾸숑
file_list = os.listdir(data_path) 
lbls = []
for i in file_list:
    number = i[:7]
    #index = phenotype_real.index[phenotype_real["eid"]==number]
    index2 = np.where(phenotype_real["eid"]==number)
    select_indices = list(index2)[0]
    select_df = phenotype_real.iloc[select_indices]
    a = list(select_df["sex"])[0]
    a = int(a)
    lbls.append(a)

lbls_df = pd.DataFrame(lbls,columns = ['sex'])
lbls_df.to_csv("npy_val_unzip_mni_100_sex.csv",index = False) #mode에 따라서 바꾸숑
```

```python
#list로 순서를 했는데 order가 중요해졌기 때문에 맞는지 확인해봄
mode = "train"
print(mode)
data_path = "/scratch/connectome/mieuxmin/UKB_t1_MNI/npy_unzip_mni_100" #mode에따라서 바꾸숑
file_list = os.listdir(data_path) 
lbls = []
name_ly = []
for i in file_list:
    number = i[:7]
    #index = phenotype_real.index[phenotype_real["eid"]==number]
    index2 = np.where(phenotype_real["eid"]==number)
    select_indices = list(index2)[0]
    select_df = phenotype_real.iloc[select_indices]
    a = list(select_df["sex"])[0]
    a = int(a)
    lbls.append(a)
    name_ly.append(number)

lbls_df = pd.DataFrame({"eid":name_ly,"sex":lbls})
#lbls_df.to_csv("test_sex.csv",index = False) #mode에 따라서 바꾸숑
```

```python
#배열 합쳐서 새롭게 저장할 거임 
path_np = "/scratch/connectome/mieuxmin/UKB_t1_MNI/npy_unzip_mni_100/"
file_list = os.listdir(path_np) 


#for i in range(len(file_list)):
 #   globals()["npy{}".format(i)] = np.load(path_np+file_list[i])
  #  globals()["npy{}".format(i)] = globals()["npy{}".format(i)].reshape(1,182,218,182)
    
#print(npy1.shape)
    
# np array설정해주기 
final = np.empty((0,1,182,218,182))
for i in range (len(file_list)):
    nmpy = np.load(path_np+file_list[i]) 
    nmpy =  nmpy.reshape(1,1,182,218,182)
    final = np.append(final, nmpy ,axis = 0)

np.save("/scratch/connectome/mieuxmin/UKB_t1_MNI/npy_unzip_mni_whole.npy",final)
print(final.shape)
```
